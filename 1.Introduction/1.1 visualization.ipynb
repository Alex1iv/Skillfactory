```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
#Надстройка %matplotlib inline сообщает Jupyter, что мы хотим получать результаты визуализаций прямо здесь, т.е. внутри файла.
```

```python
students_df = pd.read_csv('data/StudentsPerformance.csv', sep = ',')
# можно использовать ссылку на файл
#df = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv', sep = ',')

# заменим пробелы между словами в названии признаков на нижние подчеркивания. Способ 1:
students_df = students_df.rename(columns=
{'parental level of education': 'parental_level_of_education',
'test preparation course': 'test_preparation_course',
'math score': 'math_score',
'reading score': 'reading_score',
'writing score': 'writing_score'})

#или способ 2:
# students_performance.columns = [x.replace(" ", "_") for x in students_performance.columns]

# посмотрим на результат
display(students_df.head())

# типы признаков
display(students_df.dtypes)
```

```python
# диаграмма с линейной аппроксимацией взаимосвязи двух переменных
sns.lmplot(x='math_score', y='reading_score', data=students_df);
```

```python
# кастомизация графика
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_df, fit_reg = False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```

```python
"""
a Sepal width и sepal length имеют бимодальное распределение
b Petal width и petal length имеют бимодальное распределение
c Petal length имеет наибольший размах значений
d Petal width и petal length имеют унимодальное распределение
e Sepal width и sepal length имеют унимодальное распределение
"""
iris = sns.load_dataset('iris')
sns.kdeplot(data=iris);
# b,c,e
"""унимодальное - с одним горбом на линии тренда
бимодальное - с двумя
размах - дельта от минимального до максимального значения"""
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 8))
sns.kdeplot(ax=axes[0,0], x='sepal_length', data=iris)
sns.kdeplot(ax=axes[0,1], x='sepal_width', data=iris)
sns.kdeplot(ax=axes[1,0], x='petal_length', data=iris)
sns.kdeplot(ax=axes[1,1], x='petal_width', data=iris);
```

```python
fig = plt.figure(figsize=(6, 3))
ax = fig.add_axes((0, 0, 1, 1))

sns.histplot(
    data=iris,
    x='sepal_length',
    bins=30,
    kde=True,
    ax=ax
); 

# x='sepal_length', y='sepal_width', hue='petal_length', style='petal_width'
```
